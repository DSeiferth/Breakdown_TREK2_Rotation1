```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate
import time
import random
import os

import sys
path='Oscillator_Model/'
res2 = os.listdir(path)
print(res2)
sys.path.append(path)

import Model as mod
import Solution as sol
import Visualize as vis
```

```python
T2b = 5*1000
sampling_step2b = 1/1000
steps2b = int(T2b/sampling_step2b)
print('Steps: ', steps2b)

mod2b = mod.Model(gamma=0.22, driving_amplitude=2.5, driving_freq=0, initial_Condition=[0,0])
sol2b = sol.Solution(model=mod2b, tmax=T2b, nsteps=steps2b)
#sol2b.output('output2_Stochastic_driv_ampl_2-5')
```

```python
vis2b = vis.Visualize(model=mod2b, solution=sol2b)
fig2a = vis2b.plot(ratio=0.8)
fig2b = vis2b.plot_phaseplane(ratio=0.8)#pixel=0.001
fig2c = vis2b.plot_histogram()
fig2d = vis2b.plot_potential()
```

```python
path = 'output2_Stochastic_driv_ampl_2-5/'
fig2a.savefig(path+'timeseries.eps')
fig2b.savefig(path+'poincare.eps')
fig2c.savefig(path+'amplitude_distribution.eps')
fig2d.savefig(path+'potential.eps')
```
