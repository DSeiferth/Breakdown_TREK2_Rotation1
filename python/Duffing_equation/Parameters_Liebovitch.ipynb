```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate
import time
import random
import os

import sys
path='Oscillator_Model/'
res2 = os.listdir(path)
print(res2)
sys.path.append(path)

import Model as mod
import Solution as sol
import Visualize as vis
```

# Periodic driving force (Liebovitch)

```python
driving_freq=0.833
driving_period = 2 * np.pi /driving_freq
T = 100*1000#10000
sampling_step = driving_period/100#1/1000

print('steps: ', print(int(T/sampling_step)))
mod1 = mod.Model(gamma=0.22, driving_amplitude=0.16, driving_freq=driving_freq, initial_Condition=[0,0])
sol1 = sol.Solution(model=mod1, tmax=T, nsteps=int(T/sampling_step))
sol1.output('output_Liebovitch_drivPeriod_100')
```

```python
vis1 = vis.Visualize(model=mod1, solution=sol1)
fig1a = vis1.plot(ratio=0.975)
fig1b = vis1.plot_poincare(pixel=0.001)# name='test_poincare.eps',
fig1c = vis1.plot_histogram()
fig1d = vis1.plot_potential()
print('driving period = ', driving_period)
print('sampling time = ', sampling_step)
print('steps per period = ', driving_period/sampling_step)
print('number of steps sampled = %.4e' % (T/sampling_step))
```

```python
path = 'plots_Liebovitch/'
fig1a.savefig(path+'timeseries.eps')
fig1b.savefig(path+'poincare.eps')
fig1c.savefig(path+'amplitude_distribution.eps')
fig1d.savefig(path+'potential.eps')
```
